```python
import requests

response = requests.get('https://workwithus.lucioai.com/begin')
print(response.text)
```

```python
url = 'https://workwithus.lucioai.com/begin'
payload = {
    "name": "Shashwat Kumar",
    "email": "kumar.shashwat.shourya@gmail.com"
}

response = requests.post(url, json=payload)
print(response.text)
```

```python
token = response.json().get('token')
```

```python
import requests

url = 'https://workwithus.lucioai.com/get-carded'
headers = {
    "origin":  f"Bearer {token}", 
    "Authorization": f"Bearer {token}"
}

response = requests.get(url, headers=headers)
print(response.text)
```
